><h2>Importa√ß√£o de bibliotecas e Cagoriza√ß√£o do dataset</h2>

O Python como outras linguagem usa bibliotecas que cont√©m m√©todos e fun√ß√µes que podem ser reutilizadas de forma a evitar uma constante "reinven√ß√£o da roda".
Neste relat√≥rio ser√£o usadas as bibliotecas:

<ul>
    <li>Pandas: usada para manipular tabelas</li>
    <li>NumPy: usada para manipular arrays e matrizes</li>
    <li>Matplotlib:  usada para manipular e visualizar gr√°ficos</li>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

<br>Inicialmente foram criados dataframes a partir do carregamento dos ficheiros .csv<br>
<br>

```python
# Criar dataframes a partir dos ficheiros csv

atletas = pd.read_csv("athletes.csv")
jogos = pd.read_csv("sport_events.csv")

print("\nOutput:")
print(atletas.head())
print("\n")
print(jogos.head())
```

<br>As colunas das tabelas s√£o as seguintes. Este passo √© importante para termos acesso ao nome exacto de cada coluna para eventuais manipula√ß√µes.<br><br>

```python
# Informa√ß√£o sobre as tabelas (#colunas, #linhas, tipo de dados)
print("\nOutput:")
print("Colunas DataFrame atletas:", atletas.columns)
print("Colunas DataFrame jogos:", jogos.columns)
```

<br>Para a cria√ß√£o de um dataset unico √© necess√°rio juntar as tabelas. 
<br>Para isso foi utilizado o metodo .merge() do Pandas, utilizando a coluna comum √†s duas tabelas "athlete_id"<br><br>

```python
# juntar as tabelas inner-join atrav√©s de da coluna athlete_id
jogos_olimpicos = pd.merge(left=jogos, right=atletas, left_on="athlete_id", right_on="athlete_id")
print("\nOutput:")
print(jogos_olimpicos.head())
```

Neste momento exeiste um dataset √∫nico ao qual podemos verificar o nome de todas as colunas

```python
print("\nOutput:")
print("Colunas DataFrame jogos_olimpicos:", jogos_olimpicos.columns)
```

<br>A caracteriza√ß√£o dos dados √© importante porque determinados dados podem ser somados ou feitas m√©dias, como n√∫meros de identifica√ß√£o, anos de eventos...
<br>
Neste momento a caracteriza√ß√£o default dos dados √© a seguinte:
<br><br>

```python
# Informa√ß√£o sobre as tabelas (#colunas, #linhas, tipo de dados)
print("\nOutput:")
print(jogos_olimpicos.info())
```

<br>Desta forma, as colunas "year";"athlete_id";"round_id";"rank_id";"athlete_name";"date_of_birth";"country_sport_id" devem ser alteradas para dados categ√≥ricos. Dados categ√≥ricos s√£o um tipo de variavel existente na biblioteca Pandas que permite uma ordena√ß√£o dos elementos, ou seja, cada valor aquire valores que podem ser ordenados mas n√£o permite calculos estatisticos entre os elementos. 
<br><br> Por exemplo, n√£o faz sentido m√©dias dos n√∫mero de identifica√ß√£o dos atletas. O que faz sentido √© uma ordena√ß√£o, a cria√ß√£o de n√≠veis.
<br>Dessa forma √© poss√≠vel corrigir o tipo de dados com o m√©todo .astype() do pandas.
<br><br>

```python
# Corre√ß√£o de tipos de dados
# objectos/integers => categorias
jogos_olimpicos.year = jogos_olimpicos.year.astype("category")
jogos_olimpicos.athlete_id = jogos_olimpicos.athlete_id.astype("category")
jogos_olimpicos.round_id = jogos_olimpicos.round_id.astype("category")
jogos_olimpicos.rank_id = jogos_olimpicos.rank_id.astype("category")
jogos_olimpicos.athlete_name = jogos_olimpicos.athlete_name.astype("category")
jogos_olimpicos.date_of_birth = jogos_olimpicos.date_of_birth.astype("category")
jogos_olimpicos.country_sport_id = jogos_olimpicos.country_sport_id.astype("category")
```

```python
# Informa√ß√£o sobre as tabelas (#colunas, #linhas, tipo de dados)
print("\nOutput:")
print(jogos_olimpicos.info())
```

A partir deste momento j√° √© poss√≠vel utilizar o m√©todo .describe() que permite uma an√°lise geral dos dados, neste caso, o √∫nico dado para estat√≠stica √© "finish_time_seconds"

><h2>Manipula√ß√£o dos dados</h2>

Com os dados do dataset √© poss√≠vel criar novos dados, agregar e analisar com outros "olhos" todo o conjunto de dados.

<p style=text-align:justify>Ao observar os dados, percebemos que era poss√≠vel adicionar novos dados, como a velocidade da corrida, uma vez que t√≠nhamos o tamanho da pista.<br>
Para isso foi criada uma fun√ß√£o que calcula a velocidade (velocidade = dist√¢ncia / tempo) de cada jogador em cada ronda, ou seja, por cada linha da dataframe, e adiciona o valor rec√©m-calculado a uma lista.<br>Esta lista s√£o todos os valores de velocidade.</p>

```python
def column_velocity(dataframe):
    velocity = []
    lista_tempos = dataframe['finish_time_seconds'].tolist()
    for i in range(len(lista_tempos)):
        # Pista = 1500 metros
        value = 1500/lista_tempos[i]
        velocity.append(value)
    return velocity

velocidades_lista = column_velocity(jogos_olimpicos)
jogos_olimpicos['velocity (m/s)'] = velocidades_lista

print("\nOutput:")
print(jogos_olimpicos.head())
```

<br><h4>An√°lise entre anos</h4>

<p style=text-align:justify>A primeira an√°lise ser√° entre a <u><strong>contagem de atletas</strong></u> que participaram nos diferentes eventos, ser√£o contabilizados os valores √∫nicos do id do atleta.<br>
Para isso, √© aplicado √†s diferentes sub-dataframes da primeira corrida (utilizar apenas uma corrida facilita os calculos) o m√©todo .unique() (durante a itera√ß√£o, este m√©todo sinaliza o elemento com True os que s√£o √∫nicos e False os n√£o unicos), associado ao m√©todo .tolist() de forma a criar uma lista de verdadeiros, e a esta lista aplico a fun√ß√£o len(), permitindo conhecer o n√∫medo de valores unicos.
<br>
Assim, √© possivel criar um dicion√°rio associando o ano (key) √† contagem de valores √∫nicos (values).</p>
<br>

```python
# Cria√ß√£o de um dicion√°rio de contagem de atletas qe participaram no evento
nr_atletas_ano_dic = {'2002': len(round1_2002['athlete_id'].unique().tolist()),\
        '2006': len(round1_2006['athlete_id'].unique().tolist()),\
        '2010': len(round1_2010['athlete_id'].unique().tolist()),\
        '2014': len(round1_2014['athlete_id'].unique().tolist()),\
        '2018': len(round1_2018['athlete_id'].unique().tolist())}

print("\nOutput:")
print(nr_atletas_ano_dic)
```

```python
# Cria√ß√£o de dicion√°rio de convers√£o em dataframe da contagem dos atletas de cada pa√≠s por cada ano
atletas_paises_ano_dic = {'2002': round1_2002['country_sport_id'].value_counts(),\
        '2006': round1_2006['country_sport_id'].value_counts(),\
        '2010': round1_2010['country_sport_id'].value_counts(),\
        '2014': round1_2014['country_sport_id'].value_counts(),\
        '2018': round1_2018['country_sport_id'].value_counts()}

atletas_paises_ano = pd.DataFrame(atletas_paises_ano_dic)

print("\nOutput:")
print(atletas_paises_ano)
```

<h4>An√°lise entre corridas</h4>

<p>√â poss√≠vel fazermos uma an√°lise do tempo gasto por corrida e cada ano. Ou seja, verificar as vari√¢ncias entre corridas de cada ano.</p>

```python
# Agrupamento dos valores de ano e corrida com o tempo gasto de corrida
atletas_ano_tempo_medio = jogos_olimpicos.groupby(by=["year", "round_id"])["finish_time_seconds"].mean()
print("\nOutput:")
print(atletas_ano_tempo_medio)
```

```python
# Representa√ß√£o gr√°fica dos valores grupamento dos valores de tempos de corridas entre anos
atletas_ano_tempo_medio.unstack().plot(kind='pie', subplots=True, figsize=(25,5), title='Acumulado de tempos de corridas entre anos')
plt.savefig('tempos_corridas_anos.png')
```

Tamb√©m √© possivel fazer a an√°lise agrupando os valores de ano e corrida com a velocidade atingida.

```python
# Agrupamento dos valores de ano e corrida com a velocidade da corrida
atletas_ano_velocidade_media = jogos_olimpicos.groupby(by=["year", "round_id"])["velocity (m/s)"].mean()
print("\nOutput:")
print(atletas_ano_velocidade_media)
```

```python
# Representa√ß√£o gr√°fica dos valores de velocidades entre diferentes anos - visuliza√ß√£o gr√°fica n√£o sobreposta
atletas_ano_velocidade_media.unstack().plot(kind='barh', figsize=(25,5), title='Acumulado de velocidades entre anos')
plt.savefig('velocidade_corridas_anos_barh.png')
```

```python
# Representa√ß√£o gr√°fica dos valores de velocidades entre diferentes anos - visuliza√ß√£o gr√°fica sobreposta
atletas_ano_velocidade_media.unstack().plot(kind='line', figsize=(25,5), title='Acumulado de velocidades entre anos')
plt.savefig('velocidade_corridas_anos_line.png')
```

><h2>Testes Unit√°rios</h2>

Para percebermos se o c√≥digo est√° a funcionar da forma que idealizamos, √© importante criar testes unit√°rios. <br>
Assim, pensamos em alguns testes que podem ser representativos do programa geral.<br>
<ul>Casos de teste:
    <li>Exist√™ncia do ficheiro <i>athletes.csv</i> </li>
    <li>Exist√™ncia do ficheiro <i>sport_events.csv</i> </li>
    <li>Teste √† fun√ß√£o column_velocity() - dataframe de uma linha</li>
    <li>Teste √† fun√ß√£o column_velocity() - dataframe de tr√™s linhas</li>
    <li>Teste √† fun√ß√£o column_velocity() - colunas duplicadas e mesmo valores</li>
    <li>Teste √† fun√ß√£o column_velocity() - colunas duplicadas e diferentes valores</li>
    </ul>

```python
import unittest
from analise import column_velocity
import os
import pandas as pd


class MyTest(unittest.TestCase):
    
    def test_athlete_file_exist(self):
        print('\nTesting if athlete.csv file exists...')
        self.assertTrue(os.path.exists('athletes.csv'))
        print("PASS")
    
    def test_sport_events_file_exists(self):
        print('Testing if sport_events.csv file exists...')
        self.assertTrue(os.path.exists('sport_events.csv'))
        print("PASS")
    
    def test_with_1_rows(self):
        dataframe = pd.DataFrame([{0: '0', 'd': 50, 'finish_time_seconds': 5}])
        print('Testing column_velocity with 1 row...')
        self.assertEqual(column_velocity(dataframe),[300])
        print("PASS")

    def test_with_3_rows(self):
        dataframe = pd.DataFrame([{0: '0', 'd': 50, 'finish_time_seconds': 5},\
                                         {1: '1', 'd': 100, 'finish_time_seconds': 5},\
                                         {2: '2', 'd': 100, 'finish_time_seconds': 10}])
        print('Testing column_velocity with 3 rows...')
        self.assertEqual(column_velocity(dataframe),[300,300,150])
        print("PASS")

    
    def test_with_2_duplicate_columns_same_values(self):
        dataframe = pd.DataFrame([{0: '0', 'd': 50, 'finish_time_seconds': 5,'finish_time_seconds': 5},\
                                         {1: '1', 'd': 100, 'finish_time_seconds': 5, 'finish_time_seconds': 5},\
                                         {2: '2', 'd': 100, 'finish_time_seconds': 10, 'finish_time_seconds': 10}])
        print('Testing column_velocity result with duplicate columns with same values...')
        self.assertEqual(column_velocity(dataframe), [300,300,150])
        print("PASS")
        
    def test_with_2_duplicate_columns_different_values(self):
        dataframe = pd.DataFrame([{0: '0', 'd': 50, 'finish_time_seconds': 5,'finish_time_seconds': 5},\
                                         {1: '1', 'd': 100, 'finish_time_seconds': 5, 'finish_time_seconds': 10},\
                                         {2: '2', 'd': 100, 'finish_time_seconds': 10, 'finish_time_seconds': 10}])
        print('Testing column_velocity result with duplicate columns with different values...')
        self.assertEqual(column_velocity(dataframe), [300,150,150])
        print("PASS")
        
if __name__ == '__main__':
    unittest.main()
```
